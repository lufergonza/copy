####**Evaluación de los Modelos**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

def evaluar_modelo(modelo, X_train, X_test, y_train, y_test, nombre):

    y_train_pred = modelo.predict(X_train)
    y_test_pred = modelo.predict(X_test)

    print(f"\n===== {nombre} =====")

    print("\n--- MÉTRICAS TEST ---")
    print("Accuracy:", accuracy_score(y_test, y_test_pred))
    print("Precision:", precision_score(y_test, y_test_pred))
    print("Recall:", recall_score(y_test, y_test_pred))
    print("F1-score:", f1_score(y_test, y_test_pred))

    print("\nMatriz de Confusión:")
    print(confusion_matrix(y_test, y_test_pred))

    print("\nReporte Completo:")
    print(classification_report(y_test, y_test_pred))

    print("\n--- COMPARACIÓN TRAIN vs TEST (para detectar overfitting) ---")
    print("Accuracy Train:", accuracy_score(y_train, y_train_pred))
    print("Accuracy Test:", accuracy_score(y_test, y_test_pred))
```

```python
evaluar_modelo(dummy, X_train, X_test, y_train, y_test, "Dummy")
evaluar_modelo(modelo_arbol, X_train, X_test, y_train, y_test, "Decision Tree")
evaluar_modelo(knn, X_train_normalizado_imputed, X_test_normalizado_imputed, y_train, y_test, "KNN")
```

El modelo **KNN obtuvo el mejor rendimiento (0.768)**, superando tanto al **árbol de decisión (0.711)** como al **modelo base dummy (0.734)**. Esto indica que el modelo KNN logra capturar mejor los patrones presentes en los datos para predecir la cancelación de clientes.

El **árbol de decisión tuvo un rendimiento menor**, pero permite analizar de forma más clara **qué variables influyen en la cancelación**, gracias a la medida de **importancia de variables**.

# 2. Importancia de las variables (Árbol de decisión)

Las variables con mayor importancia en el modelo fueron:

| Variable                        | Importancia |
| ------------------------------- | ----------- |
| total                           | 0.541       |
| tenure                          | 0.186       |
| contrato Month-to-month         | 0.164       |
| gender                          | 0.040       |
| método de pago Electronic check | 0.032       |

### Interpretación

La variable **total (gasto total del cliente)** es el factor más influyente en la predicción de cancelación, representando más del **54% de la importancia del modelo**. Esto sugiere que el monto total que un cliente ha gastado en el servicio tiene una relación fuerte con la probabilidad de cancelación.

La segunda variable más importante es **tenure (tiempo que el cliente lleva con el servicio)**. Los clientes con menor tiempo de permanencia suelen tener mayor probabilidad de cancelar.

El **tipo de contrato Month-to-month (mensual)** también tiene una importancia considerable. Este tipo de contrato ofrece mayor flexibilidad, lo que facilita que los clientes cancelen el servicio en cualquier momento.

Las variables relacionadas con **métodos de pago**, especialmente **Electronic check**, también influyen en la cancelación, aunque en menor medida.

# 3. Principales factores que afectan la cancelación

A partir del análisis del modelo se identifican los siguientes factores clave:

### 1. Gasto total del cliente (Total)

Los clientes con menor gasto acumulado presentan mayor probabilidad de cancelar el servicio.

### 2. Tiempo de permanencia (Tenure)

Los clientes nuevos o con menor tiempo en la empresa tienen mayor tendencia a cancelar.

### 3. Tipo de contrato

Los clientes con contratos **Month-to-month** presentan mayor probabilidad de cancelación debido a la facilidad para finalizar el servicio.

### 4. Método de pago

Algunos métodos de pago, como **Electronic check**, parecen estar más asociados con la cancelación.

# 4. Estrategias de retención basadas en los resultados

Con base en los factores identificados, se pueden proponer las siguientes estrategias:

### Incentivar contratos de mayor duración

Ofrecer descuentos o beneficios a clientes que cambien de contratos **mensuales a contratos anuales o de dos años**, reduciendo así la probabilidad de cancelación.

### Mejorar la experiencia de pago

Analizar si ciertos métodos de pago generan fricción o problemas para los clientes y ofrecer alternativas más cómodas.

---

# 5. Conclusión

El análisis muestra que variables relacionadas con **el gasto total del cliente, el tiempo de permanencia y el tipo de contrato** son los factores más importantes en la predicción de cancelación.

Entre los modelos evaluados, **KNN obtuvo el mejor rendimiento predictivo**, mientras que el **árbol de decisión permitió identificar las variables más influyentes en el comportamiento de cancelación**.

El algoritmo K-Nearest Neighbors (KNN) no proporciona directamente una medida de importancia de las variables, a diferencia de otros modelos como los árboles de decisión o Random Forest. Esto se debe a que KNN es un algoritmo basado en distancias, donde la clasificación de una observación se realiza comparándola con las observaciones más cercanas dentro del conjunto de datos.

En este modelo, todas las variables participan en el cálculo de la distancia entre los datos, generalmente utilizando medidas como la distancia euclidiana. Por esta razón, el algoritmo no asigna pesos o coeficientes específicos a cada variable que permitan medir su importancia de manera directa.

Estos resultados pueden utilizarse para diseñar estrategias de retención enfocadas en clientes con mayor riesgo de cancelar, lo que permitiría mejorar la fidelización y reducir la pérdida de clientes.
